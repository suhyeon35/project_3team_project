```python
cbr = CatBoostRegressor(logging_level='Verbose', random_state=777, eval_metric='MAE')

cbr_grid = {'learning_rate': [1, 0.1, 0.01, 0.001],
            'n_estimators': [50, 100, 200, 500,1000],
            'max_depth': [1, 3, 5, 6, 8, 10, 50],
            'min_sample_split': [2,3,5,10]}
# Grid Search    
param_grid = {
    'learning_rate': [0.01, 0.1, 0.5],
    'depth': [4, 6, 8],
    'iterations': [500, 1000, 1500],
    'l2_leaf_reg': [1, 3, 5]
}

grid_search = GridSearchCV(estimator=cbr, param_grid=param_grid, cv=5, scoring='neg_mean_absolute_error', verbose=2)
# Grid Search 
grid_search.fit(x_train, y_train, cat_features=[''])

#    
print("Best Hyperparameters:", grid_search.best_params_)
print("Best MAE Score:", -grid_search.best_score_)

#   
best_model = grid_search.best_estimator_

#  
```

```python
#         MAE 
cv_mae_scores = -grid_search.cv_results_['mean_test_score']
mean_cv_mae = np.mean(cv_mae_scores)
print("Mean MAE from Cross Validation:", mean_cv_mae)
from sklearn.metrics import mean_absolute_error
#   
best_model = grid_search.best_estimator_
pred = best_model.predict(x_train)
mae = mean_absolute_error(y_train, pred)
print("MAE:", mae)

cbr_pred = best_model.predict(x_test, verbose=True)
```
